```python
# wap function that return whether a given year is leap or not ?
def leap_year(year):
    if year%4==0:
        if year%100==0:
            if year%400==0:
                return True
            else:
                return False
        else:
            return True
    else:
        return False
```

```python
year1=int(input("enter any number: "))
if leap_year(year1):
    print(year1,"leap year")
else:
    print(year1,"not a leap year")
```

```python
# wap fuction that return fibonacci series upto nth term?
def fibo(n):
    if n==1:
        return [0]
    elif n==2:
        return [0,1]
    fib=[0,1]
    for i in range(2,n):
        next_term=fib[-1]+fib[-2]
        fib.append(next_term)
    return fib
```

```python
fibo(7)
```

```python
# wap function that return n term of fibonacci series?
def fibo(n):
    if n==1:
        return 0
    elif n==2:
        return 1
    fib=[0,1]
    for i in range(2,n):
        next_term=fib[-1]+fib[-2]
        fib.append(next_term)
    return fib[-1]
```

```python
fibo(7)
```

```python
# wap function that return prime number between 1 and 100?
def prime(start,stop):
    l=[]
    for i in range(start,stop+1):
        if i>1:
            for j in range(2,i):
                if i%j==0:
                    break
            else:
                l.append(i)
    return l
```

```python
prime(1,20)
```
