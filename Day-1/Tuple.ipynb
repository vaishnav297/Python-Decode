```python
t=(10,1,25,100)
t1=sorted(t,reverse=True)
print(tuple(t1))
```

```python
t=(10,1,25,100)
t1=list(t)
for i in range(len(t1)):
    for j in range(i+1,len(t1)):
        if t1[j]>t1[i]:
            t1[i],t1[j]=t1[j],t1[i]
print(tuple(t1))
```

```python
# sorted(seqtype,[reverse=false])
# sorted function always return a new list
t=(10,1,25,100)
t1=sorted(t,reverse=True)
print(tuple(t1))
```

```python
t=(10,1,25,100)
l=list(t)
for i in range(len(l)):
    for j in range(i+1,len(l)):
        if l[i]<l[j]:
            l[i],l[j]=l[j],l[i]
print(tuple(l))
```

```python
# index
# tupleobj.index(element,[start],[stop])
t=(10,20,30,"hello",10.5,10,20,"decode")
print(t.index(20))
print(t.index(20,2))
print(t.index(20,2,6))
```

```python
# count()- occurence elments
# tupleobj.count(element)
t=(10,20,30,"hello",10.5,10,20,"decode")
print(t.count("hello"))
print(t.count(10))
```
