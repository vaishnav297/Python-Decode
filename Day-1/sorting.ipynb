```python
# index
# listobj.index(value,[start],[stop])
# by default - 0  stop - by default - end of the list
l=[10,20,30,"hello",10,10.5,1000,10,20,30,10,1000]
print(l.index(10))
print(l.index(10,1))
print(l.index(10,1,4))
```

```python
# count
# listobj.count(element)
l=[10,20,30,"hello",10,10.5,1000,10,20,30,10,1000]
print(l.count(100))
```

```python
# reverse()
# listobj.reverse()
l=[100,10,20,9]
l.reverse()
print(l)
```

```python
l=[100,10,20,9]
for i in range(len(l)):
    for j in range(i+1,len(l)):
        l[i],l[j]=l[j],l[i]
print(l)
```

```python
# sort
# listobj.sort([reverse=False])
l=[100,10,20,9]
l.sort()
print(l)
```

```python
l=[100,10,20,9]
l.sort(reverse=True)
print(l)
```

```python
# sorted()
# sorted(iterables,[reverse=False])
l=[100,10,20,9]
l1=sorted(l)
print(l1)
 
```

```python
l=[100,10,20,9]
l1=sorted(l,reverse=True)
print(l1)
```

```python
t=(100,10,20,9)
l1=sorted(t)
print(l1)
```

```python
l=[1,12,5,7]
target=19
for i in range(len(l)):
    for j in range(i+1,len(l)):
        if l[i]+l[j]==target:
            print([i,j])
```

```python
l=[(2,5),(3,1),(1,6),(4,2),(5,4),(6,3)]
for i in range(len(l)):
    for j in range(i+1,len(l)):
        if l[i][-1]>l[j][-1]:
            l[i],l[j]=l[j],l[i]
print(l)
```

```python
l=[10,1,7,5,5,7,7,2,4,6,5,7,3,9,8,7]
l1=sorted(l)
print(l1)
if len(l)%2==0:
    median=(l1[len(l1)//2-1]+l1[len(l1)//2])/2
else:
    median=l1[len(l)//2]
print(median)
```
