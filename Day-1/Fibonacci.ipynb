```python
# wapp to check whether a given number is prime or not?
n=int(input("enter any number: "))
if n==0 or n==1:
    print(n,"not a prime")
elif n>1:
    for i in range(2,n):
        if n%i==0:
            print(n,"not a prime")
            break
    else:
        print(n,"prime number")
else:
    print(n,"not a prime number")
```

```python
#wapp to check whether a given number is an armstrong number or not?
n=int(input("enter any number: "))
a=len(str(n))
sm=0
old=n
while n!=0:
    dig=n%10
    sm=sm+(dig**a)
    n=n//10
if old==sm:
    print(old,"armstrong")
else:
    print(old,"not armstrong")
```
