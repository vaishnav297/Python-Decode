```python
#syntex
#lambda arguements:expression
#  lambda function can take any number of arguments but can 
# only can noly have on expression.
# It is used with high order function like
#apply ,map,reduce and filter.
```

```python
a= lambda x:x**3
print(a(2))
```

```python
a= lambda x,y:x*y
print(a(2,3))
```

```python
a=lambda x,y:x if x>y else y
print (a(10,100))
```

```python
a=lambda x,y,z:x if x>y else y if y>z else z 
print (a(10,100,200))
```

```python
# map
# map (function ,iterables)
l=[1,2,3,4,5]
# out-[1,8,27,64,125]
l1=[]

for i in l:
    l1.append(i**3)
print(l1)
print([i**3 for i in l])
print()
```

```python
l1=[5,1,2,3,4,6]
l2=[2,6,5,4,3,1]
#[25,1,36,81,64,6]
l=[]
for i;j in zip(l1,l2):
    l.append(i**j)
print(l)
print([i**j fro i,j in zip (l1,l2)])
print (list(map(lambda i,j:i**j,l1,l2)))
```
