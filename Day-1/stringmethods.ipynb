```python
#count() - that return occurence of a character in a given string.
# strobj.count(character,[start],[end])
s="welcome to city of lakes"
print(s.count("o"))
print(s.count("e"))
print(s.count("e",2,22))
print(s.count("e",2,5))
print(s.count("z"))
```

```python
#index() - return position of any character/substring.
#strobj.index(character/substring,[start],[end])
s="welcome to bhopal"
print(s.index("o"))
print(s.index("o",5))
print(s.index("bhopal"))
print(s.index("z"))
```

```python
#find() - return position of any character/substring.
#strobj.find(character/substring,[start],[end])
s="welcome to bhopal"
print(s.find("o"))
print(s.find("o",5))
print(s.find("bhopal"))
print(s.find("z"))
```

```python
#strip
#strobj.strip(unwantedcharacter)
s="@,$,./decodedje"
print(s.strip("@.,je/$"))
```

```python
n=5
for i in range(n):
    for j in range(n-i):
        print(" ",end=" ")
    for j in range(i+1):
        print(chr(65+j),end=" ")
    for j in range(i):
        print(chr(65+j),end=" ")
    print()
```

```python
#replace
#strobj.replace(oldvalue,newvalue,[count])
#count- by default --all occurence.
s="welcome to bhopal"
print(s.replace("bhopal","decode"))
print(s.replace("o","a",1))
```

```python
#split
#strobj.split([separator],maxsplit])
#separator - by default space ,
#maxsplit - by default all occurence
#always return list of strings.
s="welcome to city of lakes"
print(s.split())
print(s.split(" ",1))
```

```python
n=5
for i in range(n):
    for j in range(n-i-1):
        print(" ",end=" ")
    for j in range((2*i)+1):
        print(chr(65+j),end=" ")
    print()
for k in range(1,n):
    for l in range(k):
        print(" ",end=" ")
    for l in range(2*(n-k)-1):
        print(chr(65+l),end=" ")
    print()
```
