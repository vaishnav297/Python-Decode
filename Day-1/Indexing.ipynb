```python
#wapp to print number of vowels, number of consonants and number of spaces in a given string?
s="welcome to bhopal"
vowel=0
consonant=0
spaces=0
for i in range(len(s)):
    if s[i] in ("a","e","i","o","u"):
        vowel=vowel+1
    elif s[i]==" ":
        spaces=spaces+1
    else:
        consonant=consonant+1
print(vowel)
print(consonant)
print(spaces)
    
```

```python
#wapp to print reverse any given string?(take user input)
s=input()
rev=""
for i in s:
    rev=i+rev
print(rev)
```

```python
#wapp to check whether a given string is palindrome or not?
s=input()
rev=""
for i in s:
    rev=i+rev
if s==rev:
    print("palindrome")
else:
    print("not palindrome")
```

```python
s=input()
rev=""
act=s
for i in range(len(s)):
    rev=s[i]+rev
if act==rev:
    print("palindrome")
else:
    print("not palindrome")
```
