```python
#wapp to check whether a triangle is equilateral, scalene or isoceles triangle (first check triangle is valid or invalid)
side1=int(input("length of side1: "))
side2=int(input("length of side2: "))
side3=int(input("length of side3: "))
if side1+side2>side3 or side1+side3>side2 or side2+side3>side1:
    if side1==side2==side3:
        print("equilateral triangle")
    elif side1!=side2!=side3:
        print("scalene triangle")
    else:
        print("isoceles triangle")
else: 
    print("invalid triangle")
```

```python
#wapp to check whether a given year is leap year or not?
year=int(input("enter any year: "))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print(year,"century leap year")
        else:
            print(year,"century year but not a leap year")
    else:
        print(year,"leap year")
else:
    print(year,"not a leap year")
```

```python
#wapp to print sum of first n even and n odd numbers?
n=int(input("enter any numbner: "))
sum_even=0
sum_odd=0
for i in range(1,2*n+1,1):
    if i%2==0:
        sum_even=sum_even+i
    else:
        sum_odd=sum_odd+i
print("sum of even numbers: ",sum_even)
print("sum of odd numbers: ",sum_odd)
```

```python
#wapp to print count of even and odd no. between 1000 and 2000 both are inclusive.
even_count=0
odd_count=0

for i in range(1000,2001):
    if i%2==0:
        even_count+=1
    else:
        odd_count+=1
print("even_numbers= ",even_count)
print("odd_numbers= ",odd_count)
```
