```python
age=np.array([23,24,23,25,26,27,27,24,27,21,23,28,29,26,25,24,22,21,25,26,25,])
print(age)
```

```python
sns.boxplot(x=age,color="purple")
plt.show()
```

```python
print(np.mean(age))
print(np.median(age))
```

```python
Q1=np.quantile(age,0.25)
print(Q1)
```

```python
Q2=np.quantile(age,0.50)
print(Q2)
```

```python
Q3=np.quantile(age,0.75)
print(Q3)
```

```python
IQR=Q3-Q1
print(IQR)
```

```python
upper_fence=Q3+(1.5*IQR)
print(upper_fence)
```

```python
lower_fence=Q1-(1.5*IQR)
print(lower_fence)
```

```python
l=[]
for i in age:
    if i>upper_fence or i<lower_fence:
        l.append(i)
print(l)
```

```python
arr=np.array([10,2,20,11,78,43,45,23,21,3,19,21])
print(arr)
print("max value:- ",np.max(arr),np.argmax(arr))
print("min value:- ",np.min(arr),np.argmin(arr))
```

```python
arr=np.array([10,2,20,11,78,43,45,23,21,3,19,21])
print(arr)
print("sqrt:- ",np.sqrt(arr))
```

```python
arr=np.array([1,0,10,20,30,5,2])
print(arr)
print("sin:- ",np.sin(arr))
print()
print("cos:- ",np.cos(arr))
```

```python
arr=np.array([10,2,20,11,78,43,45,23,21,3,19,21])
print(arr)
print("cumulative sum:- ",np.cumsum(arr))
```
