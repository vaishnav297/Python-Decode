```python
arr1=np.array([1,2,3,4,5])
arr2=np.array([6,7,8,9,10])
print(arr1)
print(arr2)
print()
arr_new=np.stack((arr1,arr2))
print(arr_new)
arr_new_v=np.vstack((arr1,arr2))
print(arr_new_v)
arr_new_h=np.hstack((arr1,arr2))
print(arr_new_h)
arr_new_d=np.stack((arr1,arr2))
print(arr_new_d)
```
